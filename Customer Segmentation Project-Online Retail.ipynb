## K-Means Clustering

```python
# Perform K-Means clustering for customer segmentation
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
individual_clustering_score=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(rfm_scaled)
    individual_clustering_score.append(kmeans.inertia_)
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10),individual_clustering_score)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(rfm_scaled)

rfm['Cluster'] = kmeans.labels_

rfm
```

```python
# Visualize customer segments
sns.scatterplot(x='Recency', y='Frequency', hue='Cluster', data=rfm)
plt.title('Customer Segmentation')
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(rfm,kmeans.labels_)
```

## Apriori Algorithm

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
pd.set_option('display.max_rows',10)
pd.set_option('display.max_columns',10)
```

```python
df=pd.read_csv(r"D:\Road to Data Science\PROJECTS\OnlineRetail (3).csv",encoding='ISO-8859-1')
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype(str)
df=df[~df['InvoiceNo'].str.contains('C')]
df.shape
```

```python
basket=(df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).
        set_index(['InvoiceNo']))
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_set=basket.applymap(encode_units)
basket_set.head()
```

```python
frequent_items=apriori(basket_set,min_support=0.07,use_colnames=True)
rules=association_rules(frequent_items,metric='lift',min_threshold=1)
rules.head()
```

```python
rules[(rules['lift']>=6) & (rules['confidence']>=0.8)].head(10)
```
